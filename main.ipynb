```python
from model import OpinionModel
import matplotlib.pyplot as plt

model = OpinionModel(num_agents=300)

knowledge = []
commitment_threshold = []
reputation = []

for agent in model.agents:
    knowledge.append(agent.knowledge)
    commitment_threshold.append(agent.involvement_threshold)
    reputation.append(agent.reputation)

plt.hist(knowledge, bins=10, label='Knowledge', edgecolor='black', linewidth=1.2)
plt.xlim(0, 1)
plt.title("Knowledge Distribution")
plt.show()
plt.hist(commitment_threshold, bins=10, label='Commitment Threshold', edgecolor='black', linewidth=1.2)
plt.title("Commitment Threshold Distribution")
plt.xlim(0, 1)

plt.show()
plt.hist(reputation, bins=10, label='Reputation', edgecolor='black', linewidth=1.2)
plt.title("Reputation Distribution")
plt.xlim(0, 1)
plt.show()


for i in range(50):
    model.step()

results = model.datacollector.get_model_vars_dataframe()
results.plot()
plt.title("Opinion Dynamics Over Time")
plt.xlabel("Step")
plt.ylabel("Number of Agents")
plt.grid(True)
plt.show()

knowledge = []
commitment_threshold = []
reputation = []

for agent in model.agents:
    knowledge.append(agent.knowledge)
    commitment_threshold.append(agent.involvement_threshold)
    reputation.append(agent.reputation)

plt.hist(knowledge, bins=10, label='Knowledge', edgecolor='black', linewidth=1.2)
plt.xlim(0, 1)
plt.title("Knowledge Distribution")
plt.show()
plt.hist(commitment_threshold, bins=10, label='Commitment Threshold', edgecolor='black', linewidth=1.2)
plt.title("Commitment Threshold Distribution")
plt.xlim(0, 1)

plt.show()
plt.hist(reputation, bins=10, label='Reputation', edgecolor='black', linewidth=1.2)
plt.title("Reputation Distribution")
plt.xlim(0, 1)
plt.show()
```
